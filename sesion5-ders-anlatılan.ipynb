Probability sampling dediğimizde mutlaka aklımıza randomization gelsin arkadaşlar. non-probability dediğimiz zaman aklımıza arbitrary, herhangi bir randomization yok.
Probability sampling'te equal chance var. Non-probability'de equal chance yok. Popülasyonun çok önemli kısımlarından çok önemli verileri alamayabilirsin.
Probability sampling dediğimizde bunun bir statistical process olduğunu bilmeniz gerekir ama non-probability sampling dediğimiz zaman tamamen sizin analitik zekanıza yönelik daha önceden belirlediğiniz analitik kurallarınıza yönelik prosedürlerin olduğunu bilmelisiniz.
Probality sampling objektif, ama no-probabilty sampling sübjektifdir.
Probability samplingin en önemli kavramlarından bir tanesi representive olmasıdır. Yani popülasyonu tam temsil ya da tama yakın kıvamda temsil edebileceğidir. Ama non-probability de bu söz konusu değildir.

T-testi, veri kümelerinin ortalamaları arasındaki farkın anlamlı olup olmadığını test etmek için kullanılan bir istatistik testidir. Formülü aşağıdaki gibidir:

t = (x̄1 - x̄2) / √[(s1^2 / n1) + (s2^2 / n2)]

Bu formülde:

t: t-skoru olarak bilinen test istatistiğidir.
x̄1: Birinci veri kümesinin ortalamasıdır.
x̄2: İkinci veri kümesinin ortalamasıdır.
s1: Birinci veri kümesinin standart sapmasıdır.
s2: İkinci veri kümesinin standart sapmasıdır.
n1: Birinci veri kümesindeki örneklem sayısıdır.
n2: İkinci veri kümesindeki örneklem sayısıdır.
T-testinin sonucunun anlamlı olup olmadığını değerlendirmek için t-skoru ile bir kritik değer arasındaki farkı karşılaştırırız. Bu kritik değer, t-testinin önceden belirlenmiş bir güven aralığına göre hesaplanır (genellikle 95% güven aralığı kullanılır). Eğer t-skoru kritik değerden büyük ise, veri kümelerinin ortalamaları arasındaki fark anlamlıdır ve bu farkın rastgele olma olasılığı düşüktür.





Bir test istatistiğinin örnekleme dağılımı, bir popülasyondan belirli bir büyüklükteki olası tüm örneklerden hesaplanan test istatistiğinin dağılımıdır. Aynı popülasyondan tekrarlanan örnekleme altında test istatistiğinin nasıl değişebileceğini açıklar. Örneğin, bir okuldaki tüm öğrencilerin ortalama boyunu tahmin etmek istediğimizi varsayalım. Bunu, öğrencilerden bir örneklem alıp boylarını ölçerek ve ardından örneklem ortalamasını hesaplayarak yapabiliriz. Ancak bu örnek ortalama, okuldaki tüm öğrencilerin gerçek boy ortalamasına tam olarak eşit olmayacaktır. Farklı bir öğrenci örneği alırsak, farklı bir örnek ortalama elde ederiz. Örnek ortalamanın örnekleme dağılımı, okuldaki tüm öğrencilerin popülasyonundan farklı örneklemler alarak elde edebileceğimiz tüm olası örnek ortalamalarının dağılımıdır. Bu işlemi birçok kez tekrarlarsak ve her örnek için örnek ortalamasını hesaplarsak, örnek boyutu yeterince büyükse (Merkezi Limit Teoremine göre) örnek ortalamasının örnekleme dağılımı yaklaşık olarak normal olacaktır. Bu, örneklem ortalamasının muhtemelen popülasyonun gerçek ortalamasına yakın olacağı ve numune ortalamasının standart hatasının tahminin belirsizliğini ölçmek için kullanılabileceği anlamına gelir.

p değeri nedir? Neden önemlidir?
 İstatistikte bir hipotez testi gerçekleştirdiğinizde, bir p-değeri, sonuçlarınızın gücünü belirlemenize yardımcı olabilir.
P-değeri 0 ile 1 arasında bir sayıdır. Değere bağlı olarak sonuçların gücünü gösterecektir.
Yargılanmakta olan iddiaya H0 (Boş Hipotez) denir.
P değeri, H0'ın doğru olduğunu varsayarsak, test istatistiğinin gözlemlenen değeri veya daha uç bir değeri alma olasılığıdır.
Daha küçük P değerleri, H0'a karşı daha güçlü kanıtları temsil eder.
Her ne kadar p değerinin 0,05 den küçük olması tıp literatüründe “istatistiksel olarak anlamlı” kabul edilse de araştırmacının isteğine göre 0,01 olarak ta kabul edilebilir. P değeri ne kadar küçük olursa H0 hipotezini reddetmek için elimizdeki kanıt o kadar yüksek olur.

p-değeri, bir test istatistiğinin örnekleme dağılımından hesaplanan bir olasılıktır. Sıfır hipotezinin doğru olduğu varsayılarak, en az numuneden hesaplanan kadar aşırı bir test istatistiğini gözlemleme olasılığını temsil eder. p-değeri, sıfır hipotezine karşı kanıtın gücünü değerlendirmek için kullanılır. Küçük bir p-değeri (tipik olarak ≤ 0,05) sıfır hipotezine karşı güçlü kanıtları gösterirken, büyük bir p-değeri (> 0,05) sıfır hipotezine karşı zayıf kanıtları gösterir. Örneğin, bir okuldaki tüm öğrencilerin ortalama boyunun 160 cm olduğu hipotezini test ettiğimizi varsayalım. Rastgele bir öğrenci örneği alıyoruz ve boylarını ölçüyoruz ve örnek ortalamasını ve ortalamanın standart hatasını hesaplıyoruz. Daha sonra bu değerleri, yaptığımız belirli teste bağlı olarak t-istatistik veya z-istatistik gibi bir test istatistiğini hesaplamak için kullanabiliriz. p-değeri, boş hipotezin (ortalama yükseklik 160 cm'dir) doğru olduğunu varsayarak, en az bizim hesapladığımız kadar aşırı bir test istatistiğini gözlemleme olasılığıdır. p-değerinin küçük olması, sıfır hipotezi doğruysa, numune ortalamasının tesadüfen meydana gelme olasılığının düşük olduğunu gösterir ve alternatif hipotez lehine boş hipotezi reddedebiliriz (ortalama yükseklik 160 cm'ye eşit değildir) ).

P değeri bir istatistik testinin sonucunun ne kadar anlamlı olduğunu gösterir. Bir istatistik testinin sonucunun anlamlı olması, hipotezin test edildiği veri setinde rastlantısal olarak oluşmuş olma olasılığının çok düşük olduğu anlamına gelir. Bu nedenle, daha düşük bir p değeri, testin sonucunun daha gerçekçi olduğu anlamına gelir.

P değeri genellikle 0,05 veya 0,01 olarak kabul edilen bir eşik değer ile karşılaştırılır. Eğer p değeri eşik değerden daha düşükse, testin sonucu anlamlı sayılır ve hipotez reddedilir. Örneğin, eğer p değeri 0,05'ten düşükse, testin sonucu anlamlı sayılır ve hipotez reddedilir. Bu, hipotezin veri setinde rastlantısal olarak oluşmuş olasılığının %5'ten düşük olduğu anlamına gelir.

Ancak, p değerinin ne kadar düşük olması gerektiği konusunda mutabakat yoktur ve bu, araştırma alanına ve amaçlara göre değişebilir. Örneğin, klinik araştırmalarda daha yüksek bir güven düzeyi gerekebilir ve bu nedenle daha düşük bir p değeri kabul edilebilir. Bu nedenle, p değerinin ne kadar düşük olması gerektiği konusunda mutabakat yoktur ve bu, araştırma alanına ve amaçlara göre değişebilir.

Tip 1 hata, bir istatistik testinde yanlış bir şekilde hipotezi reddetme hatasıdır. Yani, testin sonucunun anlamlı olmadığı halde hipotez reddedilmiştir. Bu, testin sonucunun rastlantısal olarak oluşmuş olasılığının %5'ten düşük olduğu anlamına gelir, ancak hipotezin gerçekten doğru olduğu durumlarda da sıklıkla ortaya çıkar.

Tip 1 hata, p değerine göre değerlendirilir. P değeri, testin sonucunun rastlantısal olarak oluşmuş olasılığını gösterir ve genellikle 0,05 veya 0,01 olarak kabul edilen bir eşik değer ile karşılaştırılır. Eğer p değeri eşik değerden daha düşükse, hipotez reddedilir ve tip 1 hata yapılmış olur.

Örneğin, bir araştırmada hipotez şudur: "Bu ilaç tüm hastaların %50'sinde işe yarar." Test sonucunda p değeri 0,04 olarak bulunmuştur ve hipotez reddedilmiştir. Bu durumda, testin sonucunun rastlantısal olarak oluşmuş olasılığı %4'tür ve hipotez reddedilmiştir, ancak gerçekte ilaç tüm hastaların %50'sinde işe yarayabilir. Bu durumda tip 1 hata yapılmıştır.

Tip 1 hata yapma riski, eşik değerinin daha düşük olması ile artar. Örneğin, eşik değerinin 0,01 olması durumunda tip 1 hata yapma riski daha yüksektir. Bu nedenle, eşik değerinin seçimi önemlidir ve araştırma amacına ve güven düzeyine uygun bir şekilde seçilmelidir.






Tip 2 hata, bir istatistik testinde yanlış bir şekilde hipotezi kabul etme hatasıdır. Yani, testin sonucunun anlamlı olduğu halde hipotez kabul edilmiştir. Bu, testin sonucunun rastlantısal olarak oluşmuş olasılığının %5 veya daha yüksek olduğu anlamına gelir, ancak hipotezin gerçekten yanlış olduğu durumlarda da sıklıkla ortaya çıkar.

Tip 2 hata, p değerine göre değerlendirilir. P değeri, testin sonucunun rastlantısal olarak oluşmuş olasılığını gösterir ve genellikle 0,05 veya 0,01 olarak kabul edilen bir eşik değer ile karşılaştırılır. Eğer p değeri eşik değer veya daha yüksekse, hipotez kabul edilir ve tip 2 hata yapılmış olur.

Örneğin, bir araştırmada hipotez şudur: "Bu ilaç tüm hastaların %50'sinde işe yarar." Test sonucunda p değeri 0,06 olarak bulunmuştur ve hipotez kabul edilmiştir. Bu durumda, testin sonucunun rastlantısal olarak oluşmuş olasılığı %6'tır ve hipotez kabul edilmiştir, ancak gerçekte ilaç tüm hastaların %50'sinde işe yaramayabilir. Bu durumda tip 2 hata yapılmıştır.

Tip 2 hata yapma riski, eşik değerinin daha yüksek olması ile artar. Örneğin, eşik değerinin 0,1 olması durumunda tip 2 hata yapma riski daha yüksektir. Bu nedenle, eşik değerinin seçimi önemlidir ve araştırma amacına ve güven düzeyine uygun bir şekilde seçilmelidir.






Decision table (karar tablosu), bir karar verme sürecinde farklı seçenekleri ve bu seçeneklerin sonuçlarını gösterir. Karar tablosu, bir karar verme sürecinde farklı seçeneklerin ve bu seçeneklerin olası sonuçlarının açıkça gösterilmesini sağlar, böylece en uygun seçenek seçilebilir.

Karar tablosu, genellikle birkaç sütun ve satırdan oluşur. Sütunlar, farklı seçenekleri ve satırlar ise bu seçeneklerin olası sonuçlarını gösterir. Örneğin, bir makine üretim firması için bir karar tablosu aşağıdaki gibi olabilir:

                | Seçenek 1 | Seçenek 2 | Seçenek 3 
----------------|------------|------------|------------
Maliyet        | Düşük      | Orta       | Yüksek      
Kalite         | Yüksek      | Orta       | Düşük      
Zaman          | Yüksek      | Orta       | Düşük      


Bu karar tablosunda, farklı seçeneklerin maliyet, kalite ve zaman gibi ölçütlerine göre değerlendirildiği görülmektedir. Bu tablo sayesinde, firma en uygun seçeneği seçebilir.

Karar tablosu, bir karar verme sürecinde farklı seçeneklerin ve bu seçeneklerin olası sonuçlarının açıkça gösterilmesini sağlar ve bu nedenle çok yararlı bir araçtır. Ancak, karar tablosu oluştururken dikkat edilmesi gereken bazı noktalar vardır. Örneğin, ölçütlerin doğru seçilmesi ve değerlendirilmesi önemlidir. Ayrıca, karar tablosu oluştururken subjektif değerlendirmelere yer verilmemesi önemlidir.

Confusion matrix (karmaşıklık matrisi), bir sınıflandırma modelinin doğruluk oranını değerlendirmede kullanılan bir araçtır. Karmaşıklık matrisi, bir sınıflandırma modelinin tahminlerini gerçek sınıflarla karşılaştırarak, modelin ne kadar doğru tahmin ettiğini gösterir.

Karmaşıklık matrisi, genellikle birkaç satırdan ve sütundan oluşur ve aşağıdaki gibi görülebilir:
              | Gerçek Positif | Gerçek Negatif
--------------|----------------|----------------
Tahmin Positif| TP              | FP
Tahmin Negatif| FN              | TN

Bu matriste, TP (true positive) değeri modelin doğru bir şekilde pozitif olarak sınıflandırdığı örnekleri, FP (false positive) değeri ise yanlış bir şekilde pozitif olarak sınıflandırdığı örnekleri gösterir. FN (false negative) değeri ise yanlış bir şekilde negatif olarak sınıflandırdığı örnekleri, TN (true negative) değeri ise doğru bir şekilde negatif olarak sınıflandırdığı örnekleri gösterir.

Karmaşıklık matrisi, sınıflandırma modelinin doğruluk oranını değerlendirmede yararlı bir araçtır. Örneğin, modelin doğruluk oranı (accuracy) aşağıdaki gibi hesaplanabilir:

accuracy = (TP + TN) / (TP + TN + FP + FN)

Bu oran, modelin tahminlerinin doğruluk oranını gösterir. Ayrıca, precision (doğruluk), recall (çarpım oranı) ve F1 skoru gibi diğer ölçütler de hesaplanabilir ve modelin performansı değerlendirilebilir.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
alpha = 0.5                                            #alpha
mu0 = 10                                                # hypothesized value 
xbar = 10.5                                              # sample mean 
sigma = 1.5                                             # population standard deviation 
n = 40                                                  # sample size 
z = (xbar-mu0)/(sigma/np.sqrt(n))
z                                                       # test statistic 
```

```python
pval = 1-stats.norm.cdf(z )                                     # right tail p−value
pval                                                            # print p−value
```

```python
if pval < alpha:
    print('At {} level of significance, we can reject the null hypothesis in favor of alternative hypothesis.'.format(alpha))
else:
    print('At {} level of significance, we fail to reject the null hypothesis.'.format(alpha))
```
